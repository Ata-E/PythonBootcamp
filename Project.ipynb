```python
import csv
import time

with open("menu.txt", "w") as file:
    file.write(" Lütfen bir pizza tabanı seçiniz: \n 1:Klasik\n 2:Karışık\n 3:Türk Pizzası\n Lütfen ek malzeme seçiniz: \n 4:Zeytin\n 5:Et\n 6:Soğan\n 7:Mısır\n ")

class pizza:
  def __init__(self):
    self.desc = "Herhangi bir pizza seçmediniz."
    self.cost = 0

  def get_description(self):
    return self.desc

  def get_cost(self):
    return self.cost


class klasik(pizza):
  def __init__(self):
    self.desc = "Klasik pizza, sade İtalyan usulu margarita ve domates."
    self.cost = 70
  

class karisik(pizza):
  def __init__(self):
    self.desc = "Karışık pizza, içinde ne ararsan var."
    self.cost = 100

class turkPizza(pizza):
  def __init__(self):
    self.desc = "Türk pizzası, pide benzeri ama pizza şeklinde bol kaşar bol et."
    self.cost = 90



class malzeme:
  def __init__(self):
    self.desc = "Ek malzeme seçilmedi."
    self.pizza = pizza()
    self.cost = 0

  def get_description(self):
    return pizza.get_description(self)

  def __add__(self,other):      
    return self.cost + other.cost

class zeytin(malzeme):
  def __init__(self):
    self.desc = "Zeytin eklendi."
    self.cost = 5

class et(malzeme):
  def __init__(self):
    self.desc = "Et eklendi."
    self.cost = 15

class sogan(malzeme):
  def __init__(self):
    self.desc = "Soğan eklendi."
    self.cost = 3

class misir(malzeme):
  def __init__(self):
    self.desc = "Mısır eklendi."
    self.cost = 2



def main():
  
  print("Pizzalarımıza hoşgeldiniz.\n")

  with open("menu.txt","r") as file:
    pizzas = file.read()
    print(pizzas)

  while True:
      while True:
        command = input("İstediğiniz pizzanın numarasını seçiniz: ")

        if command == "1":
            seçilenPizza = klasik()
            print(seçilenPizza.get_description())
            break
          
        elif command == "2":
            seçilenPizza = karisik()
            print(seçilenPizza.get_description())
            break

        elif command == "3":
            seçilenPizza = turkPizza()
            print(seçilenPizza.get_description())
            break
          
        else:
            print("Yaptığınız seçim pizza tabanı değil, lütfen tekrar deneyin: ")
            print(pizzas)

      while True:
        command = input("\nİstediğiniz ek malzemenin numarasını seçiniz: ")

        if command == "4":
          seçilenMalzeme = zeytin()
          print(seçilenMalzeme.get_description())
          break
          
        elif command == "5":
          seçilenMalzeme = et()
          print(seçilenMalzeme.get_description())
          break
          
        elif command == "6":
          seçilenMalzeme = sogan()
          print(seçilenMalzeme.get_description())
          break

        elif command == "7":
          seçilenMalzeme = misir()
          print(seçilenMalzeme.get_description())
          break

        else:
          print("\nYaptığınız seçim ek malzeme değil, lütfen tekrar deneyin: \n")
          print(pizzas)
      
      toplamFiyat = seçilenMalzeme + seçilenPizza
      print(f"\nToplam fiyat: {seçilenMalzeme + seçilenPizza}₺ ")
      karar = input("\nÖdemeye geçmek için 1'e basınız. Pizzanızı değiştirmek istiyorsanız 2'ye basınız.\n")

      if karar == "1": 
        while True:
          try:
              tcno = int(input("\nÖdeme için kimlik no'nuzu giriniz: "))
              break
          except ValueError:
              print("Girdiğiniz değer sayı değil, tekrar deneyiniz.")

        while True:
          try:
              kartno = int(input("\nÖdeme için kart numaranızı giriniz."))
              break
          except ValueError:
              print("Girdiğiniz değer sayı değil, tekrar deneyiniz.")

        while True:
          try:
              kartşifre = int(input("\nÖdeme için kart şifrenizi giriniz."))
              break
          except ValueError:
              print("Girdiğiniz değer sayı değil, tekrar deneyiniz.")
        break

      elif karar == "2":
        print("Tekrar seçiminizi yapınız.\n")

  
  current_time = time.strftime("%Y-%m-%d %H:%M:%S")  
  values = [tcno, kartno, kartşifre, toplamFiyat, current_time]  

  with open("orders_database.csv", "a", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["TC Kimlik Numarası", "Kredi Kartı Numarası", "Kredi Kartı Şifresi","Toplam Tutar","Ödeme Vakti"])
    writer.writerow(values)
  
  print("\nÖdemeniz alındı. Bizi tercih ettiğiniz için teşekkür ederiz.")


main()
```
